```python
#Q1 CHOLESKY

with open('m1.txt', 'r') as f1:
    l = [[int(num) for num in line.split(',')] for line in f1 ]

with open('b1.txt', 'r') as f2:
    k = [[int(num) for num in line.split(',')] for line in f2 ]

ar.printeq(l,k,'X','Given')

#check for symmetry of matrix
ar.checksym(l)

ar.printhash()

#cholesky decomposition using library function
l,lt = ar.CholeskyD(l)

ar.printhash()
#forward subsitution
ar.printeq(l,k,'Y','For forward subsitution the given')

k1=ar.forwardsub(l,k)
print('\n After forward subsitution Matrix Y is: \n', k1)

ar.printhash()
#backward subsitution
ar.printeq(lt,k1,'X','For backward subsitution the given')

k2=ar.backwardsub(lt,k1)
print('\n After backward subsitution the answer matrix X is: \n',k2)

ar.printhash()
```

```python
#Q2 LU DECOMPOSITION

with open('m2.txt', 'r') as f1:
    lq1 = [[float(num) for num in line.split(',')] for line in f1 ]

with open('b2.txt', 'r') as f2:
    kq1 = [[float(num) for num in line.split(',')] for line in f2 ]
    
ar.printeq(lq1,kq1,'X','Given')

ar.printhash()

#for pivoting the matrix i have used a different approach than the one discussed in class
#i was reading online and i found that we can pivot the matrix by introducing a pivoting matrix
#so my approach here is a lil different than that discussed in classroom
pq1=ar.pivotmat(lq1)

#pivot matrix is also useful for pivoting the other matrix
kq1=ar.crossmat(pq1,kq1)

#finding the L and U using library function
L1,U1=ar.LUdecomposition(lq1,pq1)

ar.printhash()

#forward subsitutuion
ar.printeq(L1,kq1,'Y','For forward subsitution the given')

kq11=ar.forwardsub(L1,kq1)
print('\n After forward subsitution Matrix Y is: \n', kq11)

ar.printhash()
#backward subsitution
ar.printeq(U1,kq11,'X','For backward subsitution the given')

kq12=ar.backwardsub(U1,kq11)
print('\n After backward subsitution the answer matrix X is: \n',kq12)

ar.printhash()
```

```python
#Q3
import numpy as np

#QUESTION 3


#Q1 VERIFICATION USING GAUSS JORDAN
with open('m3.txt', 'r') as f1:
    lq1 = [[float(num) for num in line.split(',')] for line in f1 ]

b = np.array([-5/3, 2/3, 3, -4/3, -1/3, 5/3])

kq1 = [[float(num)] for num in b ]

#printing the equations
ar.printeq(lq1,kq1,'X','Given')

ar.printhash()

# Initial guess for the solution
x0 = [1,1,1,1,1,1]

# Solve using Conjugate Gradient method
x = ar.conjugate_gradient(lq1, b, x0)
print(f"Solution vector by conjugate gradient method: " )
print(x)
print()


# Calculate the inverse using the solution x

# Create an empty list to store the columns of the inverse matrix
C_inv = []

# Iterate over each column of the identity matrix
for i in range(len(b)):
    # Extract the i-th column of the identity matrix (e_i)
    I_i = np.eye(len(b))[:, i]
    
    # Solve the equation Ax = e_i using Conjugate Gradient method with the solution x
    i_col1 = ar.conjugate_gradient(lq1, I_i, x0, tol=0.0001)
    
    # Append the solution (column of the inverse matrix) to the list
    C_inv.append(i_col1)

# Stack the columns of the inverse matrix horizontally to form the complete inverse matrix
A_inv = np.column_stack(C_inv)

print("Inverse of A calculated using Conjugate Gradient method:")
print(A_inv)

ar.printhash()

#solving the system of equations using gauss jordan code
ar.gaussjordan(lq1,kq1)

ar.printhash()
```

```python
# QUESTION 4

# Define the size of the linear system
N = 50
# Initial guess for the solution
x0 = np.zeros(N)
# Generate the right-hand side vector b (identity matrix)
Ib = np.eye(N)

# Generate the given matrix A
A_ij = ar.matrix_A_ij(x0)
print("Input matrix A:")
print((A_ij))
print()

# Compute the inverse of matrix A using Conjugate Gradient method
A_inv = ar.conjugate_inv(ar.matrix_A_ij, Ib, x0)

# Display the inverse matrix A_inv
print("Inverse of matrix A:")
print((A_inv))
print()

# Plot the inverse matrix
plt.figure(figsize=(8, 6))
plt.imshow(A_inv, cmap='magma')
#plt.plot(A_ij)
plt.colorbar(label='Element Value')
plt.title('Inverse Matrix')
plt.xlabel('Column Index')
plt.ylabel('Row Index')
plt.show()

# Create a meshgrid for plotting
x = np.arange(N)
y = np.arange(N)
X, Y = np.meshgrid(x, y)

# Plot the 3D surface of the inverse matrix
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, A_inv, cmap='magma')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
ax.set_title('3D Surface Plot of Inverse Matrix A')
ax.set_xlabel('Column Index')
ax.set_ylabel('Row Index')
ax.set_zlabel('Value')
plt.show()
```
